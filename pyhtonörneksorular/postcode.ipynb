```python
with open('/Users/mehmetsenel/Desktop/Örnekler/postalcode.txt',"r", encoding="Latin-1") as file :
    txt = file.readlines()


def zip_code(number):
    set_number = list(set(str(number)))

    list1 = []
    for i in set_number:
        for j in set_number:
            for k in set_number:
                for n in set_number:
                    list1.append(i+j+k+n)

    list2 = []
    for i in list1:
        if not i.startswith("0"):
            list2.append(i)


    list3 = []
    for i in txt:
            if i[0:4] in list2:
                list3.append(i)

    list4 = []
    for i in list3:
            list4.append(i.replace("\t"," ").replace("\n"," "))

    dict1 = dict()
    for i in sorted(list4,key=str.lower) :
        if i[0:4] in dict1:
             dict1[i[0:4]] += "," + i[5:]
        else :
             dict1[i[0:4]] = i[5:]



    key_list = list(dict1.keys())
    values_list = list(dict1.values())
    for i in range(len(key_list)):
         print(key_list[i] + " - " + values_list[i])
    

```

```python
zip_code(2800)
```
